```python
CATEGORIES=['Cat','Dog']
img_size=50
training_data=[]
for category in CATEGORY:
    path=os.path.join(data,category)
    class_num=CATEGORY.index(category)
    for img in os.listdir(path):
        try:
            img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
            img_new=cv2.resize(img_array,(img_size,img_size))
            list=[img_new,class_num]
            training_data.append(list)
        except Exception as e:
            pass
        
           
                     
    
```

```python
training_data=[]
def create_training_data():
    for category in CATEGORIES:
        path=os.path.join(data,category)
        class_num=CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_array=cv2.resize(img_array,(img_size,img_size))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
create_training_data()            
    
```

```python
X=[]
Y=[]
```

```python
for features,labels in training_data:
    X.append(features)
    Y.append(labels)
X=np.array(X).reshape(-1,img_size,img_size,1)
Y=np.array(Y)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,MaxPooling2D,Activation,Conv2D
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense , Flatten,MaxPooling2D,Activation,Conv2D
```

```python
dense_layers=[0,1,2]
layer_sizes=[32,64,128]
conv_layers=[1,2,3]
```

```python
X.shape
```

```python

for dense_layer in dense_layers:
    for layer_size in layer_sizes:
        for conv_layer in conv_layers:
            log_dir = "logs/fit/" + "{} -conv -{} -layersize-{}-dense layer".format(conv_layer,layer_size,dense_layer)
            
            tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
            
            model=Sequential()
            
            


            model.add(    Conv2D(layer_size,  (3,3)  ,input_shape=X.shape[1:])     )
            model.add(Activation('relu'))
            model.add(MaxPooling2D(pool_size=(2,2)))
            
            for i in range (conv_layer-1):

                model.add(    Conv2D(layer_size,  (3,3)  )     )
                model.add(Activation('relu'))
                model.add(MaxPooling2D(pool_size=(2,2)))

            model.add(Flatten())

            for i in range (dense_layer):
                model.add(Dense(layer_size))
                model.add(Activation('relu'))
                
            model.add(Dense(1))
            model.add(Activation('sigmoid'))



            model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

            

            tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)



            model.fit(X,Y,batch_size=32,epochs=12,validation_split=0.3,callbacks=[tensorboard_callback])


        

```
